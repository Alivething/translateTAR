```python
import numpy as np
import pandas as pd
```

```python
datasetnatrain = pd.read_csv("/kaggle/input/wec-intel-rectask/drive-download-20221023T093215Z-001/x_train.csv")
datasetnatest = pd.read_csv("/kaggle/input/wec-intel-rectask/drive-download-20221023T093215Z-001/x_test.csv")
ytrain = pd.read_csv("/kaggle/input/wec-intel-rectask/drive-download-20221023T093215Z-001/y_train.csv")
datasetnatrain
```

```python
datasetnatrain.isna().sum()
datasetnatest.isna().sum()
```

```python
xtrain = pd.DataFrame(datasetnatrain.fillna(datasetnatrain['Feature 3'].mean()))
xtest = pd.DataFrame(datasetnatest.fillna(datasetnatest['Feature 3'].mean()))
```

```python
xtest.isna().sum()
```

```python
xtrain.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
xtrain = sc_x.fit_transform(xtrain) 
xtest = sc_x.transform(xtest)
```

```python
ytrain = ytrain['Labels']
```

```python
from sklearn.svm import SVC
classifier = SVC()
classifier.fit(xtrain, ytrain)
```

```python
ypred = classifier.predict(xtest)
```

```python
n = len(ypred)
```

```python
ypred = pd.DataFrame(ypred, columns=['ID' ,'Labels'], index=range(n))
```
